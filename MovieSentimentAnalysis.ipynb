# Classical Sentiment Pipeline (OOP Orchestrator)

This notebook replaces large inline experimentation blocks with a reusable package architecture.
It executes the classical pipeline (TF-IDF + LR/SVM/XGB) and exports metrics and plots to the outputs folder.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
SRC_PATH = PROJECT_ROOT / 'src'
if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

PROJECT_ROOT
```

```python
from movie_sentiment.config import DataConfig, ClassicalConfig
from movie_sentiment.pipelines.classical_pipeline import ClassicalSentimentPipeline

data_cfg = DataConfig(
    data_path=PROJECT_ROOT / 'movieReviews.csv',
    phase2_comparison_path=PROJECT_ROOT / 'phase2_model_comparison.csv',
    final_comparison_path=PROJECT_ROOT / 'final_pipeline_comparison.csv',
    output_dir=PROJECT_ROOT / 'outputs'
)
model_cfg = ClassicalConfig(random_state=42, include_xgboost=True)

pipeline = ClassicalSentimentPipeline(data_config=data_cfg, model_config=model_cfg)
results_df = pipeline.run()
results_df
```

```python
artifact_path = PROJECT_ROOT / 'outputs' / 'classical' / 'phase2_model_comparison_refactored.csv'
print(f'Artifacts exported to: {artifact_path.parent}')
print('\nTop models by macro F1:')
print(results_df[['model', 'f1', 'accuracy', 'train_seconds']].head(5).to_string(index=False))
```
