# Advanced Transformer Ensemble (OOP Orchestrator)

This notebook drives the refactored transformer stack from modular Python files.
It reproduces the advanced workflow with dataset variants, class-scheme strategy, and weighted model ensembling.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
SRC_PATH = PROJECT_ROOT / 'src'
if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

PROJECT_ROOT
```

```python
from movie_sentiment.config import DataConfig, TransformerConfig
from movie_sentiment.pipelines.transformer_pipeline import AdvancedTransformerPipeline

data_cfg = DataConfig(
    data_path=PROJECT_ROOT / 'movieReviews.csv',
    phase2_comparison_path=PROJECT_ROOT / 'phase2_model_comparison.csv',
    final_comparison_path=PROJECT_ROOT / 'final_pipeline_comparison.csv',
    output_dir=PROJECT_ROOT / 'outputs'
)
model_cfg = TransformerConfig(
    model_specs=[('distilbert-base-uncased', 0.6), ('bert-base-uncased', 0.4)],
    n_splits=3,
    epochs=1,
    batch_size=8
)

pipeline = AdvancedTransformerPipeline(data_config=data_cfg, model_config=model_cfg)
results_df = pipeline.run(variant='hybrid_augmented', schemes=['five', 'three', 'two'])
results_df
```

```python
artifact_path = PROJECT_ROOT / 'outputs' / 'transformer' / 'final_pipeline_comparison_refactored.csv'
print(f'Artifacts exported to: {artifact_path.parent}')
print('\nTop pipelines by macro F1:')
print(results_df[['pipeline', 'f1', 'accuracy', 'status']].to_string(index=False))
```
